```python
#plt.plot(x,y colour= 'colour name' ,linestyle='line_style',linewidth=value,marker='marker symbol',label = 'label name)
#plt.xlabel('text')



import matplotlib.pyplot as plt

months = [1,2,3,4]
sales = [1000,1500,1200,1800]

plt.plot(months,sales,color='blue',linestyle='--',linewidth = 2,marker= 'o',label='2025 sales data')
plt.xlabel('months')
plt.ylabel('sales per month')
plt.title(' Monthly Sales data 2025')
plt.legend(loc='upper left',fontsize=9)
plt.grid(color= 'gray',linestyle=':',linewidth=.5)
plt.xlim(1,4)
plt.ylim(0,2000)
plt.xticks([1,2,3,4],['M1','M2','M3','M4'])
plt.show()
```

```python
#plt.bar(x,height,color='color name',width = vaue,label ='label name')
import matplotlib.pyplot as plt
product = ['A','B','C','D']
sale = [1000,1500,800,1200]

plt.bar(product,sale,color='orange',label = 'Sales 2025')
plt.xlabel('Product')
plt.ylabel('sales')
plt.title('Product Saes Comparison')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
#plt.pie(value,label=label_list,color=color_list,autopct='%1.1f%%')
revenue=[6000,5500,8000,7000]
region = ['East','West','North','South']
plt.pie(revenue,labels=region,colors=['red','lightgreen','blue','yellow'],autopct='%1.1f%%')
plt.show()
```

```python
#plt.hist(data,bins=numbers_of_bins,color='colorname',edgecolor='black')
import matplotlib.pyplot as plt

scores=[45,67,88,40,55,69,80,95,60,65,78,89,98,90,30,39]

plt.hist(scores,bins=5,color='purple',edgecolor='black')
plt.xlabel('scores')
plt.ylabel('Number of Students')

plt.title('Scores Distribution')
plt.show()
```

```python
#PLT.scatter(x,y,color='color_name',marker='marker_style',label='label name')
import matplotlib.pyplot as plt

hours_studies = [1,2,3,4,5,6,7,8]
exam_scores = [50,55,60,65,70,80,82,95]

#scattel plot
plt.scatter(hours_studies,exam_scores,color='red',marker='o',label='Student Data')
plt.xlabel('hours of studies')
plt.ylabel('exam scores')
plt.title('realtionship between study time and exam score')
plt.legend()
plt.grid(True)
plt.show()
```
